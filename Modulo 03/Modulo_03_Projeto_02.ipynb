###Modelo 01

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,mean_absolute_error,accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import BaggingRegressor
import statsmodels.api as sm

X = df.drop(['SalePrice'],axis=1)
y = df['SalePrice']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)

lr = LinearRegression().fit(X_train,y_train)
X_const = sm.add_constant(X)
est = sm.OLS(y,X_const).fit()
print(est.summary())
```

Os dados acima mostra uma estimativa de predição de preço das propriedades com relação a toda a base de dados, destacas-se:
 - R² = 0.916 (Mede o quão ajustado esta o modelo para o conjunto de dados)
 - F-Statistic: 77.52 (Os dados estão variando pouco entre si)
 

```python
y_pred = lr.predict(X_test)
mod01_r2 = r2_score(y_test,y_pred)
mod01_mae = mean_absolute_error(y_test,y_pred)
print(f'R-Square: {mod01_r2:.2f}')
print(f'MAE: {mod01_mae:.2f}')
```

Com base na amostra de dados apresentada foi iniciado um estudo de predição de preço de vendas de imóveis, foi fornecido uma amostra de 1460 imóveis com seu respectivos preços de vendas, assim como, 81 atributos que definiam o imóvel, esses atributos estavam definindo desde da dimensão do terreno ate a qualidade do material ultilizado na construção.

O estudo tenta predizer o preço dos imóveis com base nesses atributos tentando correlacionar os atributos mais justificavel para aplicação de preço, visa entender se atributos como dimensões do terreno e quantidade de comodos são mais eficientes do que atributos que definem qualidade e subjetividade ao imóvel como estilo, localidade, formas de acesso etc.

Primeiramente os dados foram filtrados com objetivo de encontrar os atributos com maior importancia para a predição, assim como, idnetificar dados ambiguos ou de pouca relevancia para predição dos preços, a segunda etapa consistiu em analisar os dados nulos ou faltantes para assim poder trata-los. Após o tratamento da amostra, ela foi submetida a um modelo de Regressão Logistica Multipla, o modelo consiste em traçar uma reta entre o ponto medio das vendas divindo o conjunto em dois subconjuntos e com base no valor demilitador da reta predizer o melhor preço com base nas caracteristicas.

Como objetivo é encontrar os atributos de maior relevancia os dados iniciais foram divididos em dois grupos com base nos seus atributos.
 - Grupo Categórico: Apenas atributos que indicavam qualidade ao imóvel ex: tipo de material, estilo e localidade.
 - Grupo Numérico: Apenas atributos que indicavam valores ex: dimensões e quantidade de comodos.

O modelo de regressão foi aplicado nos dois grupos de forma individual e tambem de forma coletiva, como segue:
 - Modelo 1: Usará todos os atributos (numéricos e categoricos)
 - Modelo 2: Usará apenas atributos categoricos
 - Modelo 3: Usará os atributos categoricos que melhor correlaciona com o preço de venda.
 - Modelo 4: Usará apenas os atributos numéricos.
 - Modelo 5: Usará os atributos numéricos que melhor correlaciona com o preço de venda. 
 - Modelo 6: Usará os atributos númericos e categoricos dos modelos 3 e 5 juntos.

O estudo mostrou que os atributos qualificadores tem uma grande importancia para predição de valores dos imóveis em comparação aos dados numéricos, mostrados no Relatório OLS do modelo 02 (85% de predição) e modelo 04 (76% de predição), os modelos isolados que ultilizaram os atributos mais significativos de cada grupo teve como destaque em superioridade o modelo que usou atributos qualitativos (modelo 03) com 80% de predição contra 72% do grupo com atributos numéricos (modelo 05).
